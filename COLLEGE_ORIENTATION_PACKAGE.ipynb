## **ENGINEERING COLLEGE ORIENTATION PACKAGE**

1.   RANK PREDICTOR
2.   BRANCH SUGESTOR
3.   BRANCH ELIGIBILITY PREDICTOR






```python
import ipywidgets as widgets
from IPython.display import display, clear_output
from IPython.display import HTML

# Helper Functions
def suggest_branch(technical_interests, general_interests):
    branches = {
        'CSE': {
            'technical': ['programming', 'software development', 'algorithms', 'data structures'],
            'general': ['technology', 'innovation', 'problem solving']
        },
        'ECE': {
            'technical': ['electronics', 'communication systems', 'signal processing'],
            'general': ['gadgets', 'communication', 'automation']
        },
        'EEE': {
            'technical': ['electrical systems', 'power generation', 'circuit design'],
            'general': ['energy', 'mechanical systems', 'infrastructure']
        },
        'Civil': {
            'technical': ['structural design', 'construction', 'civil engineering'],
            'general': ['architecture', 'urban planning', 'sustainability']
        },
        'Mechanical': {
            'technical': ['mechanics', 'thermodynamics', 'machine design'],
            'general': ['automobiles', 'manufacturing', 'innovation']
        },
        'DS': {
            'technical': ['data analysis', 'machine learning', 'statistics'],
            'general': ['big data', 'business intelligence', 'problem solving']
        },
        'AI': {
            'technical': ['artificial intelligence', 'machine learning', 'deep learning'],
            'general': ['technology', 'automation', 'innovation']
        },
        'AIML': {
            'technical': ['artificial intelligence', 'machine learning', 'data science'],
            'general': ['technology', 'problem solving', 'automation']
        },
        'Automobile': {
            'technical': ['vehicle design', 'automotive systems', 'mechanics'],
            'general': ['cars', 'innovation', 'manufacturing']
        }
    }

    # Scoring system
    scores = {branch: 0 for branch in branches}

    # Calculate scores based on technical and general interests
    for branch, interests in branches.items():
        tech_score = sum(1 for interest in technical_interests if interest in interests['technical'])
        gen_score = sum(1 for interest in general_interests if interest in interests['general'])
        scores[branch] = tech_score + gen_score

    # Suggest the branch with the highest score
    suggested_branch = max(scores, key=scores.get)
    return suggested_branch, scores

def predict_rank(marks, branch, community):
    data = {
        'General': {
            'CSE': {'marks': [0, 85, 75], 'ranks': [0, 1200, 3000]},
            'ECE': {'marks': [0, 80, 70], 'ranks': [0, 1400, 3500]},
            'EEE': {'marks': [0, 75, 65], 'ranks': [0, 1600, 4000]},
            'DS': {'marks': [0, 85, 78], 'ranks': [0, 1300, 3500]},
            'AI': {'marks': [0, 84, 77], 'ranks': [0, 1400, 3600]},
            'AIML': {'marks': [0, 82, 74], 'ranks': [0, 1500, 3700]},
            'Automobile': {'marks': [0, 80, 72], 'ranks': [0, 1600, 3800]}
        },
        'OBC': {
            'CSE': {'marks': [0, 80, 70], 'ranks': [0, 1800, 4000]},
            'ECE': {'marks': [0, 75, 65], 'ranks': [0, 1900, 4200]},
            'EEE': {'marks': [0, 70, 60], 'ranks': [0, 2000, 4300]},
            'DS': {'marks': [0, 80, 72], 'ranks': [0, 1900, 4200]},
            'AI': {'marks': [0, 78, 69], 'ranks': [0, 2000, 4300]},
            'AIML': {'marks': [0, 76, 68], 'ranks': [0, 2100, 4400]},
            'Automobile': {'marks': [0, 74, 65], 'ranks': [0, 2200, 4500]}
        },
        'SC': {
            'CSE': {'marks': [85, 75, 65], 'ranks': [250, 2200, 5000]},
            'ECE': {'marks': [80, 70, 60], 'ranks': [300, 2300, 5200]},
            'EEE': {'marks': [75, 65, 55], 'ranks': [350, 2400, 5400]},
            'DS': {'marks': [82, 72, 62], 'ranks': [250, 2300, 5300]},
            'AI': {'marks': [80, 70, 60], 'ranks': [260, 2400, 5400]},
            'AIML': {'marks': [78, 68, 58], 'ranks': [280, 2500, 5500]},
            'Automobile': {'marks': [75, 65, 55], 'ranks': [300, 2600, 5600]}
        },
        'ST': {
            'CSE': {'marks': [80, 70, 60], 'ranks': [300, 2500, 5500]},
            'ECE': {'marks': [75, 65, 55], 'ranks': [350, 2600, 5700]},
            'EEE': {'marks': [70, 60, 50], 'ranks': [400, 2700, 5900]},
            'DS': {'marks': [78, 68, 58], 'ranks': [320, 2600, 5800]},
            'AI': {'marks': [75, 65, 55], 'ranks': [330, 2700, 5900]},
            'AIML': {'marks': [73, 63, 53], 'ranks': [340, 2800, 6000]},
            'Automobile': {'marks': [70, 60, 50], 'ranks': [350, 2900, 6100]}
        }
    }

    # Check if marks are less than 50 for General or OBC
    if community in ['General', 'OBC'] and marks < 50:
        return 0

    # Predict rank based on marks
    try:
        branch_data = data[community][branch]
        marks_list = branch_data['marks']
        ranks_list = branch_data['ranks']

        if marks <= min(marks_list):
            return max(ranks_list)
        if marks >= max(marks_list):
            return min(ranks_list)

        for i in range(len(marks_list) - 1):
            if marks_list[i] <= marks < marks_list[i + 1]:
                slope = (ranks_list[i + 1] - ranks_list[i]) / (marks_list[i + 1] - marks_list[i])
                return int(ranks_list[i] + slope * (marks - marks_list[i]))

    except KeyError:
        return 'Error: Invalid branch or community'

    return 'Error: Rank could not be predicted'

def eligible_branches(community, rank):
    cutoffs = {
        'General': {
            'CSE': 1200, 'ECE': 1400, 'EEE': 1600, 'DS': 1300, 'AI': 1400,
            'AIML': 1500, 'Automobile': 1600
        },
        'OBC': {
            'CSE': 1800, 'ECE': 1900, 'EEE': 2000, 'DS': 1900, 'AI': 2000,
            'AIML': 2100, 'Automobile': 2200
        },
        'SC': {
            'CSE': 2500, 'ECE': 2600, 'EEE': 2700, 'DS': 2600, 'AI': 2700,
            'AIML': 2800, 'Automobile': 2900
        },
        'ST': {
            'CSE': 3000, 'ECE': 3100, 'EEE': 3200, 'DS': 3100, 'AI': 3200,
            'AIML': 3300, 'Automobile': 3400
        }
    }

    # Return an empty list if rank is 0
    if rank == 0:
        return []

    try:
        eligible = [branch for branch, cutoff in cutoffs[community].items() if rank <= cutoff]
    except KeyError:
        return 'Error: Invalid community'

    return eligible

# Interactive Widgets
def display_widgets():
    style = {'description_width': 'initial'}
    marks_input = widgets.IntSlider(min=0, max=100, step=1, description='Marks:', style=style, layout=widgets.Layout(width='500px'))
    community_input = widgets.Dropdown(options=['General', 'OBC', 'SC', 'ST'], description='Community:', style=style, layout=widgets.Layout(width='500px'))
    technical_input = widgets.SelectMultiple(
        options=['programming', 'software development', 'algorithms', 'data structures',
                 'electronics', 'communication systems', 'signal processing', 'electrical systems',
                 'power generation', 'circuit design', 'structural design', 'construction',
                 'civil engineering', 'mechanics', 'thermodynamics', 'machine design',
                 'data analysis', 'machine learning', 'statistics', 'artificial intelligence',
                 'deep learning', 'vehicle design', 'automotive systems', 'mechanics'],
        description='Technical Interests:',
        style=style,
        layout=widgets.Layout(width='500px', height='150px')
    )
    general_input = widgets.SelectMultiple(
        options=['technology', 'innovation', 'problem solving', 'gadgets', 'communication',
                 'automation', 'energy', 'mechanical systems', 'infrastructure', 'architecture',
                 'urban planning', 'sustainability', 'automobiles', 'manufacturing', 'big data',
                 'business intelligence', 'cars'],
        description='General Interests:',
        style=style,
        layout=widgets.Layout(width='500px', height='150px')
    )

    output = widgets.Output()

    def on_button_click(b):
        with output:
            clear_output()
            marks = marks_input.value
            community = community_input.value
            technical_interests = list(technical_input.value)
            general_interests = list(general_input.value)
            suggested_branch, scores = suggest_branch(technical_interests, general_interests)
            rank = predict_rank(marks, suggested_branch, community)
            eligible = eligible_branches(community, rank)

            display(HTML(f"<h4 style='color: blue;'>Suggested Branch: {suggested_branch}</h4>"))
            display(HTML(f"<h4 style='color: green;'>Scores: {scores}</h4>"))
            display(HTML(f"<h4 style='color: red;'>Predicted Rank: {rank}</h4>"))
            if isinstance(eligible, list):
                display(HTML(f"<h4 style='color: purple;'>Eligible Branches: {', '.join(eligible)}</h4>"))
            else:
                display(HTML(f"<h4 style='color: orange;'>Error: {eligible}</h4>"))

    button = widgets.Button(description="Submit", style=style)
    button.on_click(on_button_click)

    display(marks_input, community_input, technical_input, general_input, button, output)

# Display the interactive widgets
display_widgets()
```
